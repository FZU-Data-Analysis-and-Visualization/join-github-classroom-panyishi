### 一、基础创建与查看

```python
import pandas as pd
# 1. 用字典创建 DataFrame
data_dict = {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(data_dict)

# 2. 查看前1行、形状、列名、数据类型
print("前1行：")
print(df1.head(1))
print("\n形状：", df1.shape)
print("\n列名：", df1.columns.tolist())
print("\n数据类型：")
print(df1.dtypes)

# 3. 用列表创建 DataFrame，列名设为 id、tag
data_list = [[1, "A"], [2, "B"], [3, "C"]]
df2 = pd.DataFrame(data_list, columns=['id', 'tag'])
print(df2)
```

### 二、列与行操作

```python
import pandas as pd

df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

# 1. 选取 score 和 gender 两列
df_sub_cols = df[['score', 'gender']]
print("1. 选取 score 和 gender 两列：")
print(df_sub_cols)

# 2. 选取索引 0、2 的行
df_sub_rows = df.loc[[0, 2]]
print("\n2. 选取索引 0、2 的行：")
print(df_sub_rows)

# 3. 筛选 score > 80 的行
df_filter = df[df['score'] > 80]
print("\n3. 筛选 score > 80 的行：")
print(df_filter)

# 4. 新增一列 grade：score≥90 为优秀，否则为良好
df['grade'] = df['score'].apply(lambda x: '优秀' if x >= 90 else '良好')
print("\n4. 新增 grade 列后：")
print(df)

# 5. 删除 gender 列
df = df.drop('gender', axis=1)
print("\n5. 删除 gender 列后：")
print(df)
```

### 三、缺失值

```python
import pandas as pd
import numpy as np

# 1. 创建含缺失值的 DataFrame 并检测统计
data = {"x": [1, None, 3], "y": [None, 5, 6]}
df = pd.DataFrame(data)
print("1. 原始含缺失值的 DataFrame：")
print(df)

# 检测缺失值并统计每列缺失数量
print("\n每列缺失值数量：")
print(df.isnull().sum())

# 2. 用 0 填充缺失值
df_fill = df.fillna(0)
print("\n2.1 用 0 填充缺失值后：")
print(df_fill)

# 删除含缺失值的行
df_drop = df.dropna()
print("\n2.2 删除含缺失值的行后：")
print(df_drop)
```

### 四、分组聚合

```python
import pandas as pd

# 先创建基础 DataFrame（和之前的题目保持一致）
df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

# 1. 按 gender 分组，求 score 的均值和最大值
print("1. 按 gender 分组，score 的均值和最大值：")
grouped_stats = df.groupby('gender')['score'].agg(['mean', 'max'])
print(grouped_stats)

# 2. 按 gender 分组，统计每组人数
print("\n2. 按 gender 分组，每组人数：")
grouped_count = df.groupby('gender').size()
print(grouped_count)
```

### 五、排序与去重

```python
import pandas as pd

# 先创建基础 DataFrame（和之前的题目保持一致）
df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

# 1. 按 score 升序排序
df_sorted = df.sort_values(by='score', ascending=True)
print("1. 按 score 升序排序：")
print(df_sorted)

# 2. 添加重复数据并去重
# 添加重复行（Tom, 85, 男）
df = pd.concat([df, pd.DataFrame([{'name': 'Tom', 'score': 85, 'gender': '男'}])], ignore_index=True)
print("\n添加重复行后：")
print(df)

# 去重
df_deduped = df.drop_duplicates()
print("\n去重后：")
print(df_deduped)
```

### 六、合并与拼接

```python
import pandas as pd

# 1. 横向合并（按 id 合并）
df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

print("=== 1. 横向合并（按 id） ===")
print("df1:")
print(df1)
print("\ndf2:")
print(df2)

# 按 id 合并（默认内连接）
merged_df = pd.merge(df1, df2, on="id")
print("\n合并结果：")
print(merged_df)

# 2. 纵向拼接（重置索引）
df_a = pd.DataFrame({"val": [1, 2]})
df_b = pd.DataFrame({"val": [3, 4]})

print("\n=== 2. 纵向拼接（重置索引） ===")
print("df_a:")
print(df_a)
print("\ndf_b:")
print(df_b)

# 纵向拼接并重置索引
concat_df = pd.concat([df_a, df_b], ignore_index=True)
print("\n拼接并重置索引结果：")
print(concat_df)
```

### 七、时间处理

```python
import pandas as pd

# 1. 创建时间列并转为 datetime 格式
date_list = ["2025-01-01", "2025-01-02", "2025-01-03"]
# 先转成 Series，再转 datetime，这样就有 .dt 属性了
date_series = pd.Series(date_list)
date_series = pd.to_datetime(date_series)

print("转为 datetime 格式的时间列：")
print(date_series)

# 2. 提取月份、星期
print("\n提取的月份：")
print(date_series.dt.month)

print("\n提取的星期（星期一是0，星期日是6）：")
print(date_series.dt.dayofweek)
```


### 八、高级操作

```python
import pandas as pd

# 先创建基础 DataFrame
df = pd.DataFrame({
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
})

# 1. 对 score 列做标准化（(值-均值)/标准差）
score_mean = df['score'].mean()
score_std = df['score'].std()
df['score_std'] = (df['score'] - score_mean) / score_std
print("1. score 列标准化后：")
print(df[['score', 'score_std']])

# 2. 用透视表，行是 gender，值是 score，聚合函数为中位数
pivot_table = df.pivot_table(index='gender', values='score', aggfunc='median')
print("\n2. 透视表（按 gender 分组的 score 中位数）：")
print(pivot_table)
```
